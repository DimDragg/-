```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from prophet import Prophet

from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import MinMaxScaler

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

print("Файлы в папке:", os.listdir())

df = pd.read_csv("kaggel_weather_2013_to_2024.csv")

print("\nКолонки:", df.columns)
print(df.head())

# 1. Універсальна обробка дати
date_col = df.columns[0]

df[date_col] = pd.to_datetime(df[date_col])
df = df.sort_values(date_col)
df.set_index(date_col, inplace=True)

# 2. Заповнення перепусток
df = df.fillna(method="ffill")

target_col = None

for col in df.columns:
    if "temp" in col.lower():
        target_col = col
        break

print("Целевая колонка:", target_col)

data = df[target_col]

# 3. EDA
plt.figure()
plt.plot(data)
plt.title("Температура")
plt.show()

rolling = data.rolling(30).mean()

plt.figure()
plt.plot(data, label="Реальные")
plt.plot(rolling, label="Скользящее среднее")
plt.legend()
plt.show()

plt.figure()
plt.hist(data)
plt.title("Гистограмма")
plt.show()

plt.figure()
plt.boxplot(data)
plt.title("Boxplot")
plt.show()

# 4. ACF / PACF
plot_acf(data)
plt.show()

plot_pacf(data)
plt.show()

# 5. Train/Test
train_size = int(len(data)*0.8)

train = data[:train_size]
test = data[train_size:]

# 6. ARIMA
arima = ARIMA(train, order=(5,1,0))
arima_fit = arima.fit()

arima_pred = arima_fit.forecast(len(test))

# 7. Prophet
prophet_df = data.reset_index()
prophet_df.columns = ["ds","y"]

train_prophet = prophet_df[:train_size]

model = Prophet()
model.fit(train_prophet)

future = model.make_future_dataframe(periods=len(test))
forecast = model.predict(future)

prophet_pred = forecast["yhat"][-len(test):].values

# 8. LSTM
scaler = MinMaxScaler()
scaled = scaler.fit_transform(data.values.reshape(-1,1))

def create_dataset(dataset, time_step=10):
    X, y = [], []
    for i in range(len(dataset)-time_step):
        X.append(dataset[i:(i+time_step), 0])
        y.append(dataset[i+time_step, 0])
    return np.array(X), np.array(y)

time_step = 10
X, y = create_dataset(scaled, time_step)

X_train = X[:train_size]
X_test = X[train_size:]

y_train = y[:train_size]
y_test = y[train_size:]

X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

lstm_model = Sequential()
lstm_model.add(LSTM(50))
lstm_model.add(Dense(1))

lstm_model.compile(optimizer='adam', loss='mse')
lstm_model.fit(X_train, y_train, epochs=5, batch_size=32, verbose=0)

lstm_pred = lstm_model.predict(X_test)

lstm_pred = scaler.inverse_transform(lstm_pred)
y_test_real = scaler.inverse_transform(y_test.reshape(-1,1))

min_len = min(len(test), len(lstm_pred))

# 9. Метрики
def metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mape = np.mean(np.abs((y_true - y_pred)/y_true))*100
    return mae, rmse, mape

arima_m = metrics(test.values, arima_pred)
prophet_m = metrics(test.values, prophet_pred)
lstm_m = metrics(y_test_real[:min_len], lstm_pred[:min_len])

print("\nARIMA:", arima_m)
print("Prophet:", prophet_m)
print("LSTM:", lstm_m)

# 10. Графік порівняння
plt.figure()
plt.plot(test.values, label="Реальные")
plt.plot(arima_pred, label="ARIMA")
plt.plot(prophet_pred, label="Prophet")
plt.legend()
plt.title("Сравнение моделей")
plt.show()

# 11. Таблиця результатів
results = pd.DataFrame({
    "Model": ["ARIMA","Prophet","LSTM"],
    "MAE": [arima_m[0], prophet_m[0], lstm_m[0]],
    "RMSE": [arima_m[1], prophet_m[1], lstm_m[1]],
    "MAPE": [arima_m[2], prophet_m[2], lstm_m[2]]
})

print("\nРезультаты:")
print(results)
```
