```python
import simpy

def car(env):
    print("Start at", env.now)
    yield env.timeout(3)
    print("End at", env.now)

env = simpy.Environment()
env.process(car(env))
env.run()
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def model(N, t):
    r = 0.1
    return r * N

t = np.linspace(0, 10, 100)
N0 = 100
result = odeint(model, N0, t)

plt.plot(t, result)
plt.show()
```

```python
from mesa import Agent, Model

class SimpleAgent(Agent):
    def __init__(self, model):
        super().__init__(model)

    def step(self):
        print(f"Agent {self.unique_id} step")

class SimpleModel(Model):
    def __init__(self, n):
        super().__init__()

        for _ in range(n):
            agent = SimpleAgent(self)
            self.register_agent(agent)

    def step(self):
        for agent in self.agents:
            agent.step()

model = SimpleModel(3)
model.step()
```
