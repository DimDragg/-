Чуркін Дмитро, варіант 15.

0.03+0.15=0.18, 0.025+0.015=0.04

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

birth_rate = 0.18
death_rate = 0.040
N0 = [1000]
t_span = (0, 200)

# модель
def population(t, N):
    return (birth_rate - death_rate) * N

t_eval = np.linspace(0, 200, 500)

# інтегрування
sol = solve_ivp(
    population,
    t_span,
    N0,
    method='LSODA',
    t_eval=t_eval
)

# графік
plt.plot(sol.t, sol.y[0])
plt.xlabel("Час")
plt.ylabel("Чисельність популяції")
plt.title("Експоненційний ріст популяції")
plt.show()
```
