Змінити програму так, щоб кількість серверів збільшувалась або
зменшувалась в залежності від показника завантажуваноті системи.
4. Змоделювати «аварію» в системі: значне зменшення серверів в
певний момент часу. Дослідити поведінку системи.
5. Змоделювати «атаку» на систему: значне збільшення кількості
запитів, що надходять до системи. Дослідити поведінку системи.

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

#ПАРАМЕТРИ
lambda_base = 50
mu = 30
N = 3
Nmax = 25 # + 15 від мого варіанту
Nmin = 16 # + 15 від мого варіанту

max_wait_time = 0.5  # змінна для експерименту

time_span = (0, 200)
t_eval = np.linspace(0, 200, 400)

#Інтенсивність з урахуванням "атаки"
def lambda_rate(t):
    if 50 < t < 100:
        return lambda_base * 3
    return lambda_base * (1 + 0.5 * np.sin(2 * np.pi * t / 24))

# Відмови
def rejection_rate(Q, lambda_t):
    if Q > max_wait_time * lambda_t:
        return Q - max_wait_time * lambda_t
    return 0

#Кількість серверів
def N_server(t, load_factor):
    if 50 < t < 100:
        return 1

    if load_factor > 0.8:
        return Nmax
    elif load_factor < 0.2:
        return Nmin
    else:
        return N

#Швидкість обгрунтовування
def mu_rate(load_factor):
    return mu * (1 - 0.1 * load_factor)

#Модель
def smo_model(t, y):
    Q, P, R, C = y

    lambda_t = lambda_rate(t)

    load_factor = P / (N * mu) if N * mu != 0 else 0

    N_t = N_server(t, load_factor)
    mu_t = mu_rate(load_factor)

    capacity = N_t * mu_t

    processed = min(Q, capacity)
    proc_to_completed = processed * 0.8

    rejected = rejection_rate(Q, lambda_t)

    dQdt = lambda_t - processed - rejected
    dPdt = processed - proc_to_completed
    dRdt = rejected
    dCdt = proc_to_completed

    return [dQdt, dPdt, dRdt, dCdt]

initial_conditions = [0, 0, 0, 0]

#Розвязування
sol = solve_ivp(
    smo_model,
    time_span,
    initial_conditions,
    method='LSODA',
    t_eval=t_eval
)

Q, P, R, C = sol.y

# Графік
plt.figure()
plt.plot(sol.t, Q, label="Черга Q")
plt.plot(sol.t, P, label="Обробка P")
plt.plot(sol.t, R, label="Відмови R")
plt.plot(sol.t, C, label="Оброблено C")

plt.xlabel("Час")
plt.ylabel("Кількість заявок")
plt.title("СМО: атака + аварія + адаптація")
plt.legend()
plt.grid()

plt.show()
```
